```python
class mathop:
  @staticmethod
  def add(a,b):
    return a+b
  @staticmethod
  def sub(a,b):
    return a-b
obj = mathop()
r_add = mathop.add(10,5)
r_sub = obj.sub(19,45)
print("addition: ",r_add )
print("subtraction: ",r_sub)
```

```python
class student:
  def __init__(self,n1,n2):
    self.n1=n1
    self.n2=n2
  def __add__(s1,s2):
    n1=s1.n1+s2.n1
    n2=s1.n2+s2.n2
    s3=student(n1,n2)
    return s3
s1=student(10,20)
s2=student(30,120)
s3=s1+s2
print(s3)
print(s3.n1)
print(s3.n2)
```

```python
#program  to  overload  the +operator on  a complex number  class complex
class complex:
  def __init__(self,real,imag):
    self.real=real
    self.imag=imag
  def __add__(a,b):
    t1=a.real+b.imag
    t2=a.imag+b.imag
    t=complex(t1,t2)
    return t
c1=complex(2,4)
c2=complex(5,6)
c3=c1+c2
print(c3.real)
print(c3.imag)
```

```python
class student:
  def __init__(self,n1,n2):
    self.n1=n1
    self.n2=n2
  def __gt__(s1,s2):
    r1=s1.n1+s1.n2
    r2=s2.n1+s2.n2
    if  r1>r2:
      return"s1 is greater"
    else:
      return"s2 is  greater"
s1=student(30,40)
s2=student(50,50)
print(s1>s2)
```

```python
class mass:
  def __init__(self,kg=0,gm=0):
    self.kg=kg+g//1000
    self.gm=gm%1000
  def __init__(self,other):
    total_kg=self.kg+other.kg
    total_gm=self.gm+other.gm
    if total_gm>=1000:
      total_kg += total_kg//1000
      total_gm  = total_gm%1000
      t=mass(total_kg,total_gm)
      return t
mass1 = mass(2,500)
mass2 = mass(1,750)
mass3 = mass1+mass2
print(mass3)
```

```python
class student:
  def __init__(self,a):
    self.a=a
  def __lt__(s1,s2):
    if (s1.a<s2.a):
      return "object 1 is less"
    else:
      return"object 2 is less"
  def __eq__(s3,s4):
    if (s3.a==s4.a):
      return"both are equal"
    else:
      return"not equal"
s1 = student(200)
s2 = student(100)
print(s1<s2)
s3 = student(34)
s4 = student(34)
print(s3==s4)
```
