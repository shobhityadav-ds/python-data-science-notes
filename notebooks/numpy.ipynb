```python
#where operation
import numpy as np
a = np.array([1,2,3,4,5,6])
x= np.where(a==1)
print(x)
```

```python
import numpy as np
a = np.array([1,2,3,4,5,6])
x = np.where(a%2==1)
print(x)
```

```python
#sorting array
import numpy as np
a = np.array([1,2,3,41,5,6])
print(np.sort(a))
```

```python
import numpy as np
a = np.array([1,2,23,4,55,6])
y = np.where(a==2)
print(y)
x = np.searchsorted(a,2)
print(x)
```

```python
#numpy filter array
#return only true value
import numpy as np
a = np.array([1,2,3,4,5,6])
x = a[[True,False,False,True,False,True]]
print(x)
```

```python
a = np.array([13,45,65,72])
filter_a=[]
for i in a:
  if i>40:
    filter_a.append(True)
  else:
    filter_a.append(False)
new_a = a[filter_a]
print(filter_a)
print(new_a)
```
