```python
#multilevel inheritence
class grandfather:
  def __init__(self,grandfathername):
    self.grandfathername=grandfathername
class father(grandfather):
  def __init__(self,fathername,grandfathername):
    self.fathername = fathername
    grandfather.__init__(self,grandfathername)
class son(father):
  def __init__(self,sonname,fathername,grandfathername):
    self.sonname = sonname
    father.__init__(self,fathername,grandfathername)
  def p_name(self):
    print("grandfathername: ",self.grandfathername)
    print("fathername: ",self.fathername)
    print("sonname",self.sonname)
s1 = son("prince","rampal","manipal")
print(s1.grandfathername)
s1.p_name()
```

```python
class rectangle:
  def __init__(self,l,w):
    self.l=l
    self.w=w
  def area(self):
    return(self.l*self.w)
  def perimeter(self):
    return(2*self.l+2*self.w)
class square(rectangle):
  def __init__(self,l,w):
    super().__init__(l,w)
class cube(square):
  def surface_area(self):
    face_area = super().area()
    return (face_area*6)
  def volume(self):
    self.face_area = super().area()
    return(self.face_area*self.l)
c1= cube(10,10)
print(c1.area())
print(c1.surface_area())
print(c1.volume())
```

ABSTRACTION-
......abstract class - we declare and don't implement it

```python
#abstraction
from abc import ABC,abstractmethod
class father(ABC):
  @abstractmethod #we can't add anything in abstractmethod
  def display(self):
    pass
  def show(self):
    print("concrete method") #we can't make object of astract class
class child(father):
  def display(self):
    print("i am happy")
    print("abstract method")
c1 = child()
c1.show()
c1.display()
```

```python
from abc import ABC,abstractmethod
class defence(ABC):
  @abstractmethod #we can't add anything in abstractmethod
  def area(self):
    pass
  def gun(self):
    print("AK47") #we can't make object of astract class
class army(defence):
  def area(self):
    print("land")
class navy(defence):
  def area(self):
    print("sea")
class airforce(defence):
  def area(self):
    print("air")
a1 = army()
a1.area()
a1.gun()
b1 = navy()
b1.area()
b1.gun()
c1 = airforce()
c1.area()
c1.gun()
```

```python
from abc import ABC,abstractmethod
class shape(ABC):
  @abstractmethod #we can't add anything in abstractmethod
  def area(self):
    pass
  def perimeer(self):
    pass #we can't make object of astract class
class square(shape):
  def __init__(self,side):
    self.side = side
  def area(self):
    return (self.side*self.side)
  def perimeter(self):
    return (4*self.side)
class rect(shape):
  def __init__(self,l,w):
    self.l= l
    self.w =w
  def area(self):
    return (self.l*self.w)
  def perimeter(self):
    return(2*self.l+2*self.w)
s= square(3)
print(s.area())
print(s.perimeter())
r = rect(2,3)
print(r.area())
print(r.perimeter())
```

ENCAPSULATION-
THREE TYPE OF ACCCESS MODIFIER-
.....public(every one can access)
.....protected(share in parent and child only)
.....private

```python
class student:
  _name = None
  _roll = None
  _branch = None
  def __init__(self,name,roll,branch):
    self.name = name
    self.roll = roll
    self.branch = branch
  def displayinfo(self):
    print(self.roll)
    print(self.branch)
class learn(student):
  def __init__(self,name,roll,branch):
    student.__init__(self,name,roll,branch)
  def display(self):
    print(self.name)
    self.displayinfo()
ob = learn("ram",101,"ds")
ob.display()
print(ob._name)
print(ob._roll)
print(ob._branch)
```
