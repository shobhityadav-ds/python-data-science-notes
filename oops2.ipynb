```python
def div_decorator(fun1):
  def inner(x,y):
    if y==0:
      return "give proper input"
    else:
      return fun1(x,y)
  return inner
@div_decorator
def div(a,b):
  return a/b
div(4,2)
```

```python
def rev_str(my_str):
  l = len(my_str)
  for i in range(l-1,-1,-1):
    yield my_str[i]
for char in rev_str("hello"):
  print(char)
```

```python
def my_gen(x):
  while(x>0):
    if x%2 == 0:
      yield 'even'
    else:
        yield 'odd'
    x -= 1
for i in my_gen(7):
  print(i)
```

```python
def p_name(prefix):
  print("searching prefix:{}".format(prefix))
  while True:
    name = (yield)
    if prefix in name:
      print(name)
c = p_name("dear")
c.__next__()
c.send("atul")
c.send("dear atul")
```
